```python
from collections import deque
class Graph():
  def __init__(self,edges):
    self.edges={}

    for i in edges:
      self.edges[i[0]]=[]
      self.edges[i[1]]=[]

    for i in edges:
      self.edges[i[0]].append(i[1])
      self.edges[i[1]].append(i[0])


  def add_edge(self,edge):
    self.edges[edge[0]].append(edge[1])
    self.edges[edge[1]].append(edge[0])

  def add_vertex(self,v):
    self.edges[v]=[]

```

```python
def BFS(GP):
  q=deque()
  ls=[]

  key=next(iter(GP.edges.keys()))
  q.append(key)

  Visited={}
  for i in GP.edges.keys():
    Visited[i]=False

  Visited[key]=True

  while(len(q)!=0):
    ls.append(q[0])

    for i in GP.edges[q[0]]:
      if Visited[i]==False:
        q.append(i)
        Visited[i]=True

    q.popleft()

  return ls

def DFS(node,Visited,GP):
  print(node)
  Visited[node]=True
  for i in GP.edges[node]:
      if Visited[i]==False:
        DFS(i,Visited,GP)

def RecursiveDFS(GP):
  Visited={}
  for i in GP.edges.keys():
    Visited[i]=False

  key=next(iter(GP.edges.keys()))

  DFS(key,Visited,GP)

def IterativeDFS(GP):
  Visited={}
  for i in GP.edges.keys():
    Visited[i]=False

  key=next(iter(GP.edges.keys()))
  Visited[key]=True
  st=[]
  st.append(key)

  while len(st)!=0:
    temp=st.pop()
    print(temp)

    for i in GP.edges[temp]:
      if Visited[i]==False:
        st.append(i)
        Visited[i]=True
```

```python
edges = [
    (1, 2),
    (1, 3),
    (2, 3),
    (2, 4),
    (3, 5),
    (4, 5)
]
gp=Graph(edges)
print(BFS(gp))
RecursiveDFS(gp)
```

```python
class Node():
  def __init__(self,val,left=None,right=None):
    self.val=val
    self.left=left
    self.right=right


class Tree():
  def __init__(self,root=None):
    self.root=root

  @classmethod
  def TraverseInorder(cls,Node):
    ls=[]
    if not (Node.left is None):
      tls=cls.TraverseInorder(Node.left)
      for i in tls:
        ls.append(i)

    ls.append(Node.val)

    if not (Node.right is None):
      tls=cls.TraverseInorder(Node.right)
      for i in tls:
        ls.append(i)

    return ls

  @classmethod
  def TraversePreorder(cls,Node):
    ls=[]
    ls.append(Node.val)

    if not (Node.left is None):
      tls=cls.TraversePreorder(Node.left)
      for i in tls:
        ls.append(i)

    if not (Node.right is None):
      tls=cls.TraversePreorder(Node.right)
      for i in tls:
        ls.append(i)

    return ls

  @classmethod
  def TraversePostorder(cls,Node):
    ls=[]
    if not (Node.left is None):
      tls=cls.TraversePostorder(Node.left)
      for i in tls:
        ls.append(i)

    if not (Node.right is None):
      tls=cls.TraversePostorder(Node.right)
      for i in tls:
        ls.append(i)

    ls.append(Node.val)

    return ls
```

```python
A=Node('A')
B=Node('B')
C=Node('C')
D=Node('D')
E=Node('E')
F=Node('F')
A.left=B
A.right=C
B.left=D
B.right=E
C.left=F
T=Tree(A)
T.TraverseInorder(T.root)
```

```python
def BuildTreeFromPreIn(Preorder,Inorder):
  if len(Preorder)==1:
    root=Node(Preorder[0])
    return root
  elif len(Preorder)<1:
    return None

  root=Node(Preorder[0])
  idx=-1
  for i in range(len(Inorder)):
    if Inorder[i]==Preorder[0]:
      idx=i
      break
  root.left=BuildTreeFromPreIn(Preorder[1:idx+1],Inorder[0:idx])
  root.right=BuildTreeFromPreIn(Preorder[idx+1:],Inorder[idx+1:])

  return root

def BuildTreeFromPostIn(Postorder,Inorder):
  if len(Postorder)==1:
    root=Node(Postorder[-1])
    return root
  elif len(Postorder)<1:
    return None

  root=Node(Postorder[-1])
  idx=-1
  for i in range(len(Inorder)):
    if Inorder[i]==Postorder[-1]:
      idx=i
      break
  root.left=BuildTreeFromPostIn(Postorder[0:idx],Inorder[0:idx])
  root.right=BuildTreeFromPostIn(Postorder[idx:-1],Inorder[idx+1:])

  return root
```
